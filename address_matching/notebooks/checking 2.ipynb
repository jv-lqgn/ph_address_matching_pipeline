```python
# barangay level

import pandas as pd

# --------------------------------------------------
# CONFIG: choose the validation level here
# --------------------------------------------------

LEVEL = "barangay"   # options: "region", "province", "city", "barangay"

GROUP_LEVELS = {
    "region": {
        "group_cols": ["region_code"],
        "name_col": "region_name"
    },
    "province": {
        "group_cols": ["region_code", "province_code"],
        "name_col": "province_name"
    },
    "city": {
        "group_cols": ["region_code", "province_code", "city_code"],
        "name_col": "city_name"
    },
    "barangay": {
        "group_cols": ["region_code", "province_code", "city_code", "barangay_code"],
        "name_col": "barangay_name"
    }
}

group_cols = GROUP_LEVELS[LEVEL]["group_cols"]
name_col   = GROUP_LEVELS[LEVEL]["name_col"]

# --------------------------------------------------
# STEP 1: Normalize name column (SQL: TRIM(UPPER()))
# --------------------------------------------------

df = df.copy()
df["_name_clean"] = df[name_col].astype(str).str.upper().str.strip()

# --------------------------------------------------
# STEP 2: Aggregate (SQL: GROUP BY + COUNT(DISTINCT))
# --------------------------------------------------

bad_keys = (
    df
    .groupby(group_cols)["_name_clean"]
    .nunique()
    .reset_index(name="rc")
    .query("rc > 1")
)

# --------------------------------------------------
# STEP 3: Get FULL original rows causing the issue
# --------------------------------------------------

bad_rows = df.merge(
    bad_keys[group_cols],
    on=group_cols,
    how="inner"
)

# --------------------------------------------------
# (Optional) Sort for easier review
# --------------------------------------------------

sort_cols = group_cols + ["_name_clean"]
bad_rows = bad_rows.sort_values(sort_cols)

# --------------------------------------------------
# (Optional) Drop helper column
# --------------------------------------------------

bad_rows = bad_rows.drop(columns=["_name_clean"])

# --------------------------------------------------
# OUTPUT
# --------------------------------------------------

print(f"Validation level : {LEVEL}")
print(f"Name column used : {name_col}")
print(f"Problematic groups found: {len(bad_keys)}")
print(f"Total violating rows    : {len(bad_rows)}")

# bad_keys  -> SQL-equivalent aggregated result
# bad_rows  -> FULL original rows
```

```python
import pandas as pd

# --------------------------------------------------
# CONFIG
# --------------------------------------------------

LEVEL_CONFIG = {
    "region"  : {"code_col": "region_code",  "name_col": "region_name"},
    "province": {"code_col": "province_code", "name_col": "province_name"},
    "city"    : {"code_col": "city_code",     "name_col": "city_name"},
    "barangay": {"code_col": "barangay_code", "name_col": "barangay_name"}
}

# --------------------------------------------------
# HELPER: normalize
# --------------------------------------------------

def normalize(series):
    return series.astype(str).str.strip()

# --------------------------------------------------
# HELPER: blank detection
# catches NaN, None, empty string, whitespace, "nan", "none", "null"
# --------------------------------------------------

def is_blank(series):
    s = normalize(series)
    return (
        series.isna() |
        s.eq("") |
        s.str.lower().isin(["nan", "none", "null"])
    )

# --------------------------------------------------
# MAIN: check all levels
# only flags rows where:
#   - code column is blank   AND
#   - name column has an actual entry
# --------------------------------------------------

df = df.copy()

results = {}

for LEVEL, config in LEVEL_CONFIG.items():

    code_col = config["code_col"]
    name_col = config["name_col"]

    bad_code_rows = df[
        is_blank(df[code_col]) &    # code is blank
        ~is_blank(df[name_col])     # name has an actual entry
    ].copy()

    results[LEVEL] = bad_code_rows

    print(f"{'='*50}")
    print(f"Level          : {LEVEL}")
    print(f"Code checked   : {code_col}")
    print(f"Name checked   : {name_col}")
    print(f"Bad rows found : {len(bad_code_rows)}")

print(f"{'='*50}")

# --------------------------------------------------
# EXPORT TO EXCEL
# --------------------------------------------------

OUTPUT_PATH = "blank_code_results.xlsx"

with pd.ExcelWriter(OUTPUT_PATH, engine="openpyxl") as writer:
    for LEVEL, bad_code_rows in results.items():
        bad_code_rows.to_excel(writer, sheet_name=f"{LEVEL}_bad_code", index=False)

print(f"Results saved to: {OUTPUT_PATH}")

# --------------------------------------------------
# HOW TO ACCESS IN MEMORY
# --------------------------------------------------

# results["region"]
# results["province"]
# results["city"]
# results["barangay"]
```
